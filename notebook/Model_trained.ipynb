```python
## split the data into independent and depenent variable
```

```python
X = df.drop(['SK_ID_CURR','TARGET'],axis=1)
y = df['TARGET']
X.shape, y.shape
```

```python
## feature engineeiring applying
numerical_featuare = X.select_dtypes(include=['int64','float64']).columns
categorical_feature = X.select_dtypes(include=['object']).columns

print(f"numerical feature: {list(numerical_featuare)} ")
print(f"Categorical feature: {list(categorical_feature)} ")
```

```python
from sklearn.preprocessing import StandardScaler,OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer


num_pipe = Pipeline(
    steps=[
        ("imputer",SimpleImputer(strategy="median")),
        ("scaler", StandardScaler())
    ]
)

cat_pipe = Pipeline(
    steps=[
        ("imputer", SimpleImputer(strategy="most_frequent")),
        ("one_hot_encoder",OneHotEncoder(handle_unknown="ignore"))
        
    ]
)

preprocessor = ColumnTransformer(
    [
        ("num",num_pipe,numerical_featuare),
        ("cat",cat_pipe,categorical_feature)
    ]
)
```

```python
## split data into train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.20,stratify=y,random_state=42)
X_train.shape, X_test.shape
```

```python
X_train_transformer=preprocessor.fit_transform(X_train)
X_test_transformer = preprocessor.transform(X_test)
X_train_transformer
```

```python
X_test_transformer
```

```python
## handling inbalance dataset
from imblearn.over_sampling import SMOTE
from collections import Counter

smote= SMOTE(random_state=42)

X_train_resamble, y_train_resamble=smote.fit_resample(X_train_transformer,y_train)

print(f"Distribution of data ",Counter(y_train_resamble))
```

```python
## Model evaluation metric
from sklearn.metrics import f1_score,recall_score,precision_score,accuracy_score
def evaluate_metric(true,pred):
    f1score = f1_score(true,pred)
    recall = recall_score(true,pred)
    precision = precision_score(true,pred)
    accuracy = accuracy_score(true,pred)
    return f1score,recall, precision,accuracy
```

```python
## model training
```

```python
## importing libary for model training
from sklearn.linear_model import LogisticRegressionCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier,AdaBoostClassifier,GradientBoostingClassifier
from catboost import CatBoostClassifier
from xgboost import XGBClassifier
```

```python
## initialize the model
models = {
    "LogisticRegression":LogisticRegressionCV(),
    "DecisionTreeClassifier":DecisionTreeClassifier(),
    "RandomForestClassifier": RandomForestClassifier(),
    "AdaboostClassier":AdaBoostClassifier(),
    "GradientBoostingClassifier": GradientBoostingClassifier(),
    "CatboostClassifier":CatBoostClassifier(),
    "XgboostClassifier":XGBClassifier()
}
```

```python
model_list = []
f1_score_list = []
recall_score_list = []
precision_score_list = []
accuracy_score_list = []



for model_name, model in models.items():
    model.fit(X_train_resamble,y_train_resamble)
    #y_pred = model.predict(X_test_transformer)
    y_pred = (model.predict_proba(X_test_transformer)[:,1] >= 0.5).astype(int)


    ## evaluate the model with testing dat
    f1_test_score,recall_test_score,precision_test_score,accuracy_test_score = evaluate_metric(y_test,y_pred)

    ## appending the model
    model_list.append(model_name)
    f1_score_list.append(f1_test_score)
    recall_score_list.append(recall_test_score)
    precision_score_list.append(precision_test_score)
    accuracy_score_list.append(accuracy_test_score)

    result_df = pd.DataFrame ({
        "Model":model_list,
        "F1 score":f1_score_list,
        "recall score": recall_score_list,
        "precision score":precision_score_list,
        "accuracy_score":accuracy_score_list
    })
    
```

```python
## applying hyperparameter
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param = {
    "LogisticRegression": {
        'penalty': ['l1', 'l2'],
        'solver': ['liblinear'],
        'random_state': [42]
    },
    "DecisionTreeClassifier":{
        'max_depth':[3,5,None],
        'min_samples_split':[2],
        'splitter':['best']
    },
    "RandomForestClassifier":{
        'n_estimators':[50,100,200],
        'min_samples_split':[2,4,6],
        'min_samples_leaf':[1,3,5]
    },
    "XgboostClassifier":{
        'learning_rate':[0.01,0.1,0.2],
        'max_depth':[2,3,6],
        'n_estimators':[20,50,100],
        'reg_alpha':[0.1],
        'scale_pos_weight':[1],
        'colsample_bytree':[0.8]
    }
    
    
}
```

```python
model_list_grid = []
f1_score_list_grid = []
recall_score_list_grid = []
precision_score_list_grid = []
accuracy_score_list_grid = []


for model_name, model in models.items():
    if model_name in param:
        gs = GridSearchCV(estimator = model,param_grid=param[model_name],cv=5,scoring="f1_weighted")
        gs.fit(X_train_resamble,y_train_resamble)
        best_model = gs.best_estimator_
        ## prediction
        y_pred_grid = gs.predict(X_test_transformer)
        ## evaluating model
        f1_score_test_grid,recall_score_test_grid,precision_score_test_grid,accuracy_score_test_grid = evaluate_metric(y_test,y_pred_grid)
        ## appending the model
        model_list_grid.append(model_name)
        f1_score_list_grid.append(f1_score_test_grid)
        recall_score_list_grid.append(recall_score_test_grid)
        precision_score_list_grid.append(precision_score_test_grid)
        accuracy_score_list_grid.append(accuracy_score_test_grid)

```

```python
# creating data frame
result_grid = pd.DataFrame({
    "Best Model":model_list_grid,
    "f2 score grid":f1_score_list_grid,
    "recall score grid":recall_score_list_grid,
    "precision score grid":precision_score_list_grid,
    "accuracy score grid":accuracy_score_list_grid
})
```
